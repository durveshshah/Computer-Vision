```python
# Name: Durvesh Shah
# Student Id: 1131901

import numpy as np
import cv2
from google.colab.patches import cv2_imshow

#Read and Converting to grayscale image
fish = cv2.imread('/content/fish.bmp',cv2.IMREAD_GRAYSCALE)
submarine = cv2.imread('/content/submarine.bmp',cv2.IMREAD_GRAYSCALE)

Submarine_Low_Pass_Filter = cv2.GaussianBlur(submarine,(23,23),9)
#cv2_imshow(Low_Pass_Filter)

Submarine_High_Pass_Filter = submarine - Submarine_Low_Pass_Filter
cv2_imshow(Submarine_High_Pass_Filter)

fish_Low_Pass_Filter = cv2.GaussianBlur(fish,(23,23),3)
cv2_imshow(fish_Low_Pass_Filter)

Hybrid_Image =  fish_Low_Pass_Filter + Submarine_High_Pass_Filter
cv2_imshow(Hybrid_Image)

''' As observed during trial and error, low pass filter generally has a low sigma value.
     In this case, when sigma increases i.e. 9 we look at a larger picture and we don't tend to look at all the details.
     However, when the sigma decreases i.e. 3, we tend to get more detailed picture.  '''
```
