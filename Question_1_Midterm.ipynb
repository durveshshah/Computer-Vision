```python
# Importing important libraries
import cv2
from scipy import ndimage
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
# Showing and plotting original image
original_image = cv2.imread('/content/plane.bmp')
# Changing image to 1D array for displaying histogram and image
one_d_array_1 = original_image.mean(axis=2).flatten()

# Plotting Histogram
hist = plt.hist(one_d_array_1)
plt.title("Histogram of Original Image")
plt.show()
print("Original Image")
cv2_imshow(original_image)

```

```python
# rotating an image to 17 degrees
rotated_image = ndimage.rotate(original_image, 17)
# Changing image to 1D array for displaying histogram and image
one_d_array_2 = rotated_image.mean(axis=2).flatten()
plt.title("Histogram of Rotated Image to 17 degree")

# Plotting Histogram
plt.hist(one_d_array_2)
plt.show()
print("Rotated Image 17 degree")
cv2_imshow(rotated_image)
```

```python
# Translating an image
# Moving 1 pixel to the right and 5 pixels up
tanslated_image = T = np.float32([[1, 0, 25], [0, 1, -5]])
# wrapAffine in cv2 translates the image into the given array above
changed_image = cv2.warpAffine(original_image, tanslated_image, (original_image.shape[1], original_image.shape[0]))
# Changing image to 1D array for displaying histogram and image
one_d_array_4 = changed_image.mean(axis=2).flatten()

# Plotting Histogram
plt.hist(one_d_array_4)
plt.title("Histogram of Translated Image")
plt.show()
print("Translated Image")
cv2_imshow(changed_image)
```

```python
#Scaling an image
scaled_image = cv2.resize(original_image, (450, 390))
# Changing image to 1D array for displaying histogram and image
one_d_array_3 = scaled_image.mean(axis=2).flatten()

# Plotting Histogram
plt.hist(one_d_array_3)
plt.title("Histogram of Scaled Image")
plt.show()
print("Scaled Image")
cv2_imshow(scaled_image)
```

```python
''' 1. Taken input from original image and put as scaled image input
    2. Taken output of scaled image and put as input in rotated image
    3. Taken output of rotated image and put as input in translated image'''
fig,ax0 = plt.subplots()
first_image = cv2.imread('/content/plane.bmp')

# Scaling the image
scaled_image_new = cv2.resize(first_image, (450, 390))

# Rotating the image to 17 degrees
rotated_image_new = ndimage.rotate(scaled_image_new, 17)

# Translating the image. Moving 25 pixels to the right and 5 pixels up
tanslated_image_new = T = np.float32([[1, 0, 25], [0, 1, -5]])
# wrapAffine in cv2 translates the image into the given array above
changed_image_new = cv2.warpAffine(rotated_image_new, tanslated_image_new, (rotated_image_new.shape[1], rotated_image_new.shape[0]))
# Changing images to 1D array for displaying histogram and image
one_d_array_new_1 = changed_image_new.mean(axis=2).flatten()
one_d_array_new_2 = rotated_image_new.mean(axis=2).flatten()
one_d_array_new_3 = scaled_image_new.mean(axis=2).flatten()

# Plotting Histogram
ax0.hist(one_d_array_new_1,label = 'Translated Image')
ax0.hist(one_d_array_new_2,label = 'Rotated Image')
ax0.hist(one_d_array_new_3,label = 'Scaled Image')
ax0.legend()
plt.title("Histogram of Scaled, Rotated and Translated Image")
plt.show()
print("Final Output of all 3 images")
cv2_imshow(changed_image_new)
```
