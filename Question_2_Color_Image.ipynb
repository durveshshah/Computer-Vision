```python
# Name: Durvesh Shah
# Student Id: 1131901

import numpy as np
import cv2
from google.colab.patches import cv2_imshow

#Read and Converting to grayscale image
fish = cv2.imread('/content/fish.bmp')
submarine = cv2.imread('/content/submarine.bmp')

fish = cv2.cvtColor(fish, cv2.COLOR_BGR2RGB)
submarine = cv2.cvtColor(fish, cv2.COLOR_BGR2RGB)

Submarine_Low_Pass_Filter = cv2.GaussianBlur(submarine,(23,23),29)
#cv2_imshow(Low_Pass_Filter)

Submarine_High_Pass_Filter = submarine - Submarine_Low_Pass_Filter
cv2_imshow(Submarine_High_Pass_Filter)

fish_Low_Pass_Filter = cv2.GaussianBlur(fish,(23,23),17)
cv2_imshow(fish_Low_Pass_Filter)

Hybrid_Image =  fish_Low_Pass_Filter + Submarine_High_Pass_Filter
cv2_imshow(Hybrid_Image)

''' As observed in BGR2RGB image, even if I change sigma values in low pass filters it does not tend to produce hybrid images  '''
```
