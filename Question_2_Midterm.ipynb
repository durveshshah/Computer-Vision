```python
import cv2 
from google.colab.patches import cv2_imshow

source_image = cv2.imread('/content/Fruits_examimage.jpg') 

# Resizing the image to smaller resolution to detect the fruits
source_image = cv2.resize(source_image,(320, 200))
cv2_imshow(source_image)

# Converting to grayscale image.
grayscale = cv2.cvtColor(source_image, cv2.COLOR_BGR2GRAY) 
cv2_imshow(grayscale)

# Converting to blurred image so that we can perform Hough transform.
blurred_image = cv2.medianBlur(grayscale,7)
cv2_imshow(blurred_image)

# Calling openCV hough circles library and passing the parameters.
''' 
Parameters used: 

1. Passing the 8 bit input blurred image

2. Passing OpenCV Hough Gradient method

3. We have put 1 as 1 is the dp of the image. We can increase or decrease the dp or we can say adjust the resolution

4. The fourth parameter is the least distance between the focuses of the detected circles. 
if the boundary is too little, numerous neighbor circles might be dishonestly distinguished missing a genuine one. 
In the event that it is too huge, a few circles might be missed.

5. The fifth parameter s param1 which is used in canny edge detector.

6. The sixth one is param2 which is used for detecting circles. The lower the value, the more false circles are detected.

7. The seventh one is the minimum radius of the circle.

8. The eigth one is the maximum radius of the circle.

 '''
circle_shape = cv2.HoughCircles(blurred_image,cv2.HOUGH_GRADIENT, 1, source_image.shape[0]/16, param1=212, param2=10, minRadius=10, maxRadius=35)
count = 0

for i in circle_shape[0,:]:
    cv2.circle(source_image,(i[0],i[1]),i[2],(244,255,4),4)
    count+=1
cv2_imshow(source_image)
print()
print("Number of fruits detected is approximately: ",count)


''' 
References
1. Reference taken from Processing_3 file provided in Colab Handouts.
2. https://docs.opencv.org/4.5.2/d3/de5/tutorial_js_houghcircles.html (USing Hough Circle parameters)
'''
```
